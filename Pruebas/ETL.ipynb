### Carga y etl Checkin de Yelp 

```python
# Cargo df checkin de yelp
data = []
with open("C:/Users/54280/Downloads/Yelp/checkin.json", "r") as file:
    for line in file:
        data.append(json.loads(line))

df_checkin_yelp = pd.DataFrame(data)
```

```python
df_checkin_yelp
```

```python
# Chequeo valores nulos
df_checkin_yelp.isnull().sum()
```

```python
# Chequeo df
df_checkin_yelp.info()
```

```python
# Chequeo valores duplicados
df_checkin_yelp.duplicated().sum()
```

```python
# Genero una funcion para dividir las fechas

desanidar = []

for _, row in df_checkin_yelp.iterrows():
    business_id = row['business_id']
    dates = row['date'].split(', ')
    for date in dates:
        time, date = date.split(' ')
        desanidar.append([business_id, time, date])

df_desa_checkin_yelp = pd.DataFrame(desanidar, columns=['business_id', 'hour', 'fecha'])
```

```python
# Chequeo como quedo el df 
df_desa_checkin_yelp
```

```python
# Exporto df desanidado
df_desa_checkin_yelp.to_parquet('df_checkin_yelp')
```
